### Naver 금융 스크래핑

```python
import requests, bs4, pandas as pd, numpy as np
import chardet

resp = requests.get("http://finance.naver.com/marketindex/")

# 인코딩 방식 설정
# resp.encoding='utf-8'
resp.encoding='euc-kr'
html = resp.text

bs = bs4.BeautifulSoup(html,'html.parser')
 
# bs4모듈의 select_one 함수는 DOM 트리구조를 정확하게 입력해야 사용가능.
# span 의 태그에 띄어쓰기가 존재한다면 반드시 띄어쓰기 동일하게 해야 함
titles = bs.select("h3.h_lst > span.blind  ")
# type(nations)  
  
    
nameList = []
for title in titles:
    textedTitle = title.text
    nameList.append(textedTitle)
# nameList

prices = bs.select("div.head_info > span.value")
# price

# 고시가의 쉼표 제거
quotList = []
for price in prices:
    textedMoney = price.text
    rmDeliMny =  textedMoney.replace(",","")
    quotList.append(rmDeliMny)
quotList

# 컬럼명 삽입은 항상 리스트형으로 처리
data_name = pd.DataFrame(nameList,columns=['주요지표'])
# # # data_name
data_qout = pd.DataFrame(quotList, columns=['환율&시세'])
# # # data_qout
mergedData = pd.concat([data_name,data_qout], axis=1)
mergedData

# # # # CSV 파일로 저장, encondig 한글깨짐 방지
# mergedData.to_csv("../dataset/fxData.csv", encoding='utf-8-sig',index=False)
# mergedData

# engineMyPc = create_engine('oracle+cx_oracle://hyo:hyo2@192.168.110.2:1522/xe')
# resultname ='FINANCE_NAVER_정효재'
# mergedData.to_sql(resultname, engineMyPc, if_exists='replace', index=False)
```

### Daum 파이낸스 스크래핑 : select -> find

```python
import requests, bs4, pandas as pd, numpy as np
import chardet

resp = requests.get("http://finance.daum.net/quote/marketvalue.daum")

# 인코딩 방식 설정
resp.encoding='utf-8'
# resp.encoding='euc-kr'
html = resp.text
bs = bs4.BeautifulSoup(html,'html.parser')
 
# stock_list = bs.find("ul",{"id":"myListTop1"}).findAll("li")    
# stock_list

stocks = bs.select("ul#myListTop1 > li")
# stocks

stockCompany = []
stockPrice = []
count = 0
totalNumbers = 6
for stock in stocks:
    if count > totalNumbers:
        break
    #--------------------------------------------------  
    stock_Company = stock.find("a").text
    stockCompany.append(stock_Company)
    #--------------------------------------------------    
    stock_Price = stock.find("span",{"class":"cost num"}).text.replace(",","")
    stockPrice.append(stock_Price)
    count += 1

TopPrice   = pd.DataFrame(stockCompany, columns=['Company'], dtype=str)    
TopCompany = pd.DataFrame(stockPrice, columns=['Price'], dtype=float)

# TopPrice.head()
# TopCompany.head()

final = pd.concat([TopPrice,TopCompany],axis =1 )
final

 
# final.to_csv("../dataset/fxData.csv", encoding='utf-8-sig',index=False)
# final

# engineMyPc = create_engine('oracle+cx_oracle://hyo:hyo2@192.168.110.2:1522/xe')
# resultname ='FINANCE_DAUM_정효재'
# final.to_sql(resultname, engineMyPc, if_exists='replace', index=False)
```

### Daum 파이낸스 스크래핑 : select -> select_one

```python
import requests, bs4, pandas as pd, numpy as np
from sqlalchemy import create_engine 

resp = requests.get("http://finance.daum.net/quote/marketvalue.daum")
resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')
 
stocks1 = bs.select("ul#myListTop1 > li")
stockName=[]
stockPrice=[]
num = 0 
numbers = 6

for stock in stocks:
    if num > numbers:
        break
    #--------------------------------------------------
    #stock_price = stock1.find("span",{"class":"cost num"}).text.replace(",","") 
    stock_price = stock.select_one("span.cost").text.replace(",","")
    stockPrice.append(stock_price)
    #--------------------------------------------------
    #stock_name = stock1.find("a").text    
    stock_name = stock.select_one("a").text
    stockName.append(stock_name)
    
    num += 1

TopPrice1   = pd.DataFrame(stockPrice, columns=['Price'], dtype = float)    
TopCompany1 = pd.DataFrame(stockName, columns=['Company'], dtype = str) 
final = pd.concat([TopCompany1,TopPrice1],axis =1 )
final


# final.to_csv("../dataset/fxData.csv", encoding='utf-8-sig',index=False)
# final

# engineMyPc = create_engine('oracle+cx_oracle://hyo:hyo2@192.168.110.2:1522/xe')
# resultname ='FINANCE_DAUM_정효재'
# final.to_sql(resultname, engineMyPc, if_exists='replace', index=False)
```

### Daum 파이낸스 주가동향 스크래핑 : 컬럼 , 데이터 분리된 스크래핑

```python
import requests, bs4, pandas as pd, numpy as np
from sqlalchemy import create_engine 

resp = requests.get("http://finance.daum.net/quote/marketvalue.daum")
resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')
# print(bs)

stockQuots = bs.findAll("span",{"class":"cost num"})
stockQuots
# stockName
stocks = bs.find("ul",{"id":"myListTop1"})
stocksName = stocks.findAll("a")
# type(stocksName)

count=0
totalCount=9
companies = list()
for name in stocksName:
    if count > totalCount:
        break
    company = name.text
    companies.append(company)
    count += 1   
companies       

count=0
cNumber=9
quotations = []
for quot in stockQuots:    
    textedQ = quot.text.replace(",","")
    if count > cNumber:
        break
    quotations.append(textedQ)
    count += 1
quotations    

title1 =["회사명"] 
title2 = ["시가"]

df1 = pd.DataFrame(data=companies, columns=title1)
df2 = pd.DataFrame(data=quotations, columns=title2)
resultFrame = pd.concat([df1,df2], axis=1)
resultFrame

# resultFrame.to_csv("../dataset/fxData.csv", encoding='utf-8-sig',index=False)
# resultFrame

# engineMyPc = create_engine('oracle+cx_oracle://hyo:hyo2@192.168.110.2:1522/xe')
# resultname ='FINANCE_DAUM_정효재'
# resultFrame.to_sql(resultname, engineMyPc, if_exists='replace', index=False)
```

### Daum 주가 동향 스크래핑 2 : double for loop && Dict to DF

```python
import requests, bs4, pandas as pd, numpy as np
from sqlalchemy import create_engine 

resp = requests.get("http://finance.daum.net/quote/marketvalue.daum")
resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')


count=0
nameNumber=9
count1=0
priceNumber=9
companies = list()
quotations = []


stockCollcet = bs.select("ul#myListTop1 > li ")
for stock in stockCollcet:
    allA = stock.select("a")
    for a in allA:
        cName = a.text
        if count > nameNumber:
            break
        companies.append(cName) 
        count+=1            
    allCost = stock.findAll("span",{"class":"cost num"})     
    for cost in allCost:
        cPrice = cost.text.replace(",","")
        if count1 > priceNumber:
            break          
        quotations.append(cPrice)    
        count1+=1
        
# companies
# quotations

#  딕셔너리를 이용해 데이터 프레임 생성
testData  = pd.DataFrame\
            (
            {'companies':companies,
            'quotations':quotations}
            )

print(testData)


# testData.to_csv("../dataset/FINANCE_DAUM.csv", encoding='utf-8-sig',index=False)
# testData

# engineMyPc = create_engine('oracle+cx_oracle://hyo:hyo2@192.168.110.2:1522/xe')
# resultname ='FINANCE_DAUM_정효재'
# testData.to_sql(resultname, engineMyPc, if_exists='replace', index=False)
```
