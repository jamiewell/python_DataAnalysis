### 이상데이터 정제

```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

 
engineORC_0 = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl')

customerData = pd.read_sql_query(""" SELECT * FROM customerdata """, engineORC_0) 
customerData.head()
```

#### 데이터프레임[컬럼]  = np.where(조건식, true, false(=>else) )

```python
step = list(range(1,4))

# 기호 \ = 개행
customerData["productage_age_new"] = np.where(customerData["productage"] < step[0], step[0],\
                                            np.where(customerData["productage"] < step[1], step[1],\
                                            np.where(customerData["productage"] < step[2], step[2],\
                                            5)))

customerData.head() 
```

```python
customerData["productage_age_new_mix"] = np.where(customerData["avgprice"] >4000 , 'high',\
                                                  np.where(customerData["avgprice"] > 3400 , 'middle',\
                                                           'below'))


# np.min(customerData.avgprice),\
# np.percentile(customerData.avgprice,25),\
# np.median(customerData.avgprice),\
# np.percentile(customerData.avgprice,50),\
# np.percentile(customerData.avgprice,75),\
# np.max(customerData.avgprice)

customerData.head()
```

#### 함수로 정제

```python
def refining(df):
    if df["productage"] < 1:
        return 1
    if df["productage"] < 2:
        return 2
    if df["productage"] < 3:
        return 3
    else:
        return 5        
```

```python
customerData["productage_new2"] = customerData.apply(refining, axis=1)
customerData.head()
```

### 이동평균

```python
# 이동평균법 계산이 가능한 rolling 함수 // spark의 sliding과 같음
customerData["TEST"]  = customerData["productage"].rolling(window = 5, center = True).mean()
customerData.head()
```

```python
# suborder = 이동평균계산 상 맨 앞, 맨 뒤에 커버가 안되는 영역
import math
suborder = math.floor(5/2)
# suborder

listFront = []
for i in range(0,suborder):
    listFront.append(customerData["productage"][0:i+suborder].mean())
listFront    


totalSize = len(customerData["productage"])


listBack = []
for i in range(2):
    listBack.append(customerData["productage"][148-i:totalSize].mean())    
listBack 
```

```python
for i in range (0,suborder):
    customerData["TEST"][i] = listFront[i]
    customerData["TEST"][148+i] = listBack[i]
customerData.tail()
```

```python
# 147 , 148 ,149 /3
std148 = customerData["productage"][147:150].mean()

# 148 ,149 /2
std149 = customerData["productage"][148:150].mean()

n = repr(std148)
n2 = repr(std149)
print(n + '   ' + n2)
```

#### 데이터 합치기

```python
import pandas as pd
from sqlalchemy import create_engine

 
engineORC_0 = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl')

# showTabs = pd.read_sql_query(""" SELECT * FROM  tabs where 1=1 and table_name like '%CUSTOMERDATA%'  """, engineORC_0) 

regionMasterData = pd.read_sql_query(""" SELECT * FROM  KOPO_REGION_MST """, engineORC_0) 
selectExample = pd.read_sql_query(""" SELECT * FROM  KOPO_CHANNEL_SEASONALITY_NEW """, engineORC_0) 
```

#### 데이터 테이블 조인

```python
leftJoinTable = pd.merge(selectExample,regionMasterData,\
        left_on='regionid', right_on='regionid',how="left", indicator=True)\
[['regionid','regionname','product','yearweek','qty']]
leftJoinTable
```

#### 비율 계산

```python
meanValue = leftJoinTable.\
groupby(['regionid','product'],as_index=False)[['qty']].\
mean().rename(columns={'qty':'avg_qty'})

mergeTable = pd.merge(selectExample,meanValue,\
 
mergeTable['ratio'] = selectExample['qty'] / mergeTable['avg_qty']

mergeTable
```

#### 이동평균 계산

```python
def sub_function(data):
    data['ma2'] = mergeTable["qty"].rolling(window = 5, center = True).mean()
    return data
```

#### 집계테이블에서 이동평균 계산

```python
groupResult = mergeTable.groupby(["regionid","product"]).apply(sub_function)
groupResult

# groupResult.to_csv("../dataset/groupResult_test1.csv", index=False)
```
